```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_num = 1000000
sec_num = 1000
```

```python
p = np.random.uniform(0, 1, (sample_num, ))
x = np.random.normal(-3, 4, (sample_num, ))
y = np.random.normal(2, 4, (sample_num, ))
```

```python
z = p * x + (1-p) * y
```

```python
z_min = min(z)
z_max = max(z)
sec = np.linspace(z_min, z_max, num=sec_num)
pdf = np.histogram(z, bins=sec_num)[0]/len(z)
cdf = np.cumsum(pdf, axis=0)
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("Z")
ax1.set_ylabel("PDF", color="red")
ax1.plot(sec, pdf, color="red")
ax1.tick_params(axis="y", labelcolor="red")
ax2 = ax1.twinx()
ax2.set_ylabel("CDF", color="green")
ax2.plot(sec, cdf, color="green")
ax2.tick_params(axis="y", labelcolor="green")
fig.tight_layout()
plt.show()
```

```python
quantile = sec[np.min(np.where(cdf>0.9))]
print(quantile)
```

```python
expect = np.average(sec, weights=pdf)
variance = np.var(z)
print(expect, variance)
```
